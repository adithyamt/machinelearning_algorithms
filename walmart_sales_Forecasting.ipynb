```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
feature=pd.read_csv('features.csv')
store=pd.read_csv('stores.csv')
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
train=train.groupby(['Store','Date'])['Weekly_Sales'].sum()
train=train.reset_index()
train.head(5)
```

```python
df=pd.merge(train,feature,on=['Store','Date'],how='inner')
df.head(15)
```

```python
df=pd.merge(df,store,on=['Store'],how='inner')
df.head(5)
```

```python
df=df.sort_values(by='Date')
df.head(5)
```

```python
sns.countplot('Type',data=df)
```

```python
sns.countplot('IsHoliday',data=df)
```

```python
df["Weekly_Sales"].plot.hist()
```

```python
df.isnull().sum()
```

```python
df=df.drop(['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5'],axis=1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['IsHoliday']=[int(i) for i in list(df.IsHoliday)]
```

```python
df.head()
```

```python
dummy_type=pd.get_dummies(df['Type'],drop_first=True)
dummy_type.head()
```

```python
df=pd.concat([df,dummy_type],axis=1)
df.head()
```

```python
df=df.drop(['Type'],axis=1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x=df.drop(['Weekly_Sales','Store','Date'],axis=1)
y=df['Weekly_Sales']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression(normalize=True)
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
```

```python
plt.figure(figsize=(20,15))
plt.plot(y_test,prediction,'ro')
plt.plot(y_test,y_test,'b-')
plt.show()
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,prediction)
print(r2)
```
